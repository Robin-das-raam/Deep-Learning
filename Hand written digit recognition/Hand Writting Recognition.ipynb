# Prepare Dataset for training

```python
print(X_train.shape)
print(X_train[0].shape[0])
print(X_train[0].shape[1])
```

```python
img_rows = X_train[0].shape[0]
img_cols = X_train[1].shape[0]


X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)

# store the shape of a single image

input_shape = (img_rows, img_cols, 1)

print(input_shape,"input_shape")
```

```python
#Change the image type into float data type

X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
```

```python
#normalize the data by changeing the range from (0 to 255)

X_train /= 255.0
X_test /= 255.0
```

```python
print("X_train Shape", X_train.shape)
print("X_test Shape", X_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

print("y_test shape", y_test.shape)
print("y_train shape", y_train.shape)
```

```python
X_train.shape[1] * X_train.shape[2]
```

```python
print("number of classes ", str(y_test.shape[1]))

y_train[0]
```

```python
num_class = y_test.shape[1]
num_pixels = X_train.shape[1] * X_train.shape[2]
```

# Creat our model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from tensorflow.keras import backend as K


model = Sequential()

model.add(Conv2D(32, kernel_size = (3,3),
                 activation = "relu",
                 input_shape = input_shape))

model.add(Conv2D(64, (3,3),activation = "relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation = "relu"))
model.add(Dropout(0.50))
model.add(Dense(num_class,activation = "softmax"))

model.compile(loss = "categorical_crossentropy",
              optimizer = SGD(0.01),
              metrics = ["accuracy"])

print(model.summary())
```

# Train our Model

```python
batch_size = 32
epochs = 5

history = model.fit(X_train,y_train,
                    epochs = epochs,
                    batch_size = batch_size,
                    verbose = 1,
                    validation_data = (X_test, y_test))

score = model.evaluate(X_test, y_test, verbose = 0)
print("Test loss ", score[0])
print("Test accuracy ",score[1])
```

# Plotting accuracy and Loss

```python
history.history
```

```python
##plotting our loss Chart

import matplotlib.pyplot as plt

history_dict = history.history

loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]

epochs = range(1, len(loss_values)+1)

line1 = plt.plot(epochs, val_loss_values, label = "validation/Test_loss")
line2 = plt.plot(epochs , loss_values, label = "Training_loss")

plt.setp(line1, linewidth = 2.0, marker = "*", markersize = 10.0)
plt.setp(line2, linewidth = 2.0 , marker = "o", markersize = 10.0)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()
plt.show()
```

```python
##plotting our accuracy Chart

acc_values = history_dict["accuracy"]
val_acc_values = history_dict["val_accuracy"]

epochs = range(1, len(acc_values)+1)

line1 = plt.plot(epochs, acc_values, label= "Training_accuracy")
line2 = plt.plot(epochs ,val_acc_values, label = "Validation/Test_accuracy")

plt.setp(line1, linewidth = 2.0, marker = "o", markersize = 10.0)
plt.setp(line2, linewidth = 2.0, marker = "*", markersize = 10.0)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()
plt.show()
```
